# I have json and csv data of USA youtube data 

```python
import pandas as pd  

df1=pd.read_json(r"C:\Users\balwa\Downloads\youtube_dataset\US_category_id.json")
df=pd.read_csv(r"C:\Users\balwa\Downloads\youtube_dataset\USvideos.csv")
```

```python
import pandas as pd

# Creating df_categories DataFrame with category information
df_categories = pd.DataFrame([
    {'category_id': i, 'category_name': item['snippet']['title']}
    for i, item in enumerate(df1['items'])  # Loop through items and extract category information
])

# Join df_categories with df based on 'category_id'
data = df_categories.join(df.set_index('category_id'), on='category_id')  # Join category information with the main dataframe

# Creating utube DataFrame as a copy of 'data'
utube = data.copy()

# Handling missing values
utube.dropna(subset=['channel_title'], inplace=True)  # Drop rows with missing 'channel_title'

# Remove unwanted columns from the dataframe
unwanted_columns = ['video_id', 'thumbnail_link', 'video_error_or_removed', 'description']
utube.drop(unwanted_columns, axis=1, inplace=True)

# Convert date columns to datetime format
utube['publish_time'] = pd.to_datetime(utube['publish_time'])  # Convert 'publish_time' to datetime
utube['trending_date'] = pd.to_datetime(utube['trending_date'], format='%y.%d.%m')  # Convert 'trending_date' to datetime

# Extract and format 'publish_month', 'publish_day', and 'publish_year'
utube['publish_month'] = utube['publish_time'].dt.strftime('%B')
utube['publish_day'] = utube['publish_time'].dt.day
utube['publish_year'] = utube['publish_time'].dt.year

# Convert 'views' to float and round
utube['views'] = utube['views'].astype(float).round(2)

# Calculate metrics grouped by category and channel_title
metrics = utube.groupby(['category_name', 'channel_title']).agg({
    'views': 'mean',
    'likes': 'mean',
    'dislikes': 'mean',
    'comment_count': 'mean',
    'publish_year': 'mean'
}).sort_values(by=['views', 'likes', 'dislikes', 'comment_count'])

# Rename columns
column_name = {
    'views': 'avg_views',
    'likes': 'avg_likes',
    'dislikes': 'avg_dislikes',
    'comment_count': 'avg_comment_count',
    'publish_year': 'avg_publish_year'
}
metrics.rename(columns=column_name, inplace=True)

# Display the calculated metrics further you can visualize the data
display(metrics)
```
